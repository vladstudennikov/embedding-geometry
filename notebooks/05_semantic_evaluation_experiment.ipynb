# Semantic Evaluation of Clusters

This notebook performs semantic analysis on the results of the `ClusterAssignmentExperiment`. For each clustering run, it identifies:
- **Core Words**: Top TF-IDF keywords that define the cluster.
- **Cluster Size**: Number of points in each cluster.
- **Unique/Strange Words**: Keywords that are highly specific to a single cluster.

```python
import sys
import pandas as pd
sys.path.append("..")
```

```python
from src.experiments.semantic_evaluation import SemanticEvaluationExperiment
from src.utils.config import load_all_configs
```

```python
# Load config to get the correct text column name
config = load_all_configs("dimension_sweep")
text_column = config["dataset"]["text_column"]
print(f"Analyzing semantics for column: {text_column}")
```

```python
# Initialize and run experiment
experiment = SemanticEvaluationExperiment(
    source_experiment="cluster_assignment_sweep",
    experiment_name="semantic_evaluation"
)
experiment.run(text_column=text_column)
```

```python
# Example of loading one report
import json
from pathlib import Path

reports = list(experiment.results_dir.glob("*.json"))
if reports:
    with open(reports[0], "r", encoding="utf-8") as f:
        sample_report = json.load(f)
    
    print(f"Sample results from {reports[0].name}:")
    for cid, data in list(sample_report.items())[:3]:
        print(f"\nCluster {cid} ({data['n_points']} points):")
        print(f"  Core words: {', '.join(data['top_keywords'][:5])}")
        print(f"  Unique words: {', '.join(data['unique_keywords'][:5])}")
```
