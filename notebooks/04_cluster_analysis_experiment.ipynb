# Cluster Analysis Experiment

This notebook analyzes the results generated by the `ClusterAssignmentExperiment`. It calculates key statistics for each clustering run and visualizes how these metrics change across UMAP dimensions and algorithms.

```python
import sys
import pandas as pd
sys.path.append("..")
```

```python
from src.experiments.cluster_analysis import ClusterAnalysisExperiment
from src.visualization.analysis_plots import AnalysisPlotter
```

```python
# Initialize and run analysis
experiment = ClusterAnalysisExperiment(
    source_experiment="cluster_assignment_sweep", 
    experiment_name="cluster_analysis"
)
summary_df = experiment.run()
```

```python
# Display first few rows of the summary
if summary_df is not None:
    display(summary_df.head(10))
```

```python
if summary_df is not None:
    plotter = AnalysisPlotter(experiment.figures_dir)
    print("Generating plots...")
    plotter.plot_metric_vs_dimension(
        summary_df,
        metric="n_clusters"
    )
    plotter.plot_metric_vs_dimension(
        summary_df,
        metric="normalized_entropy"
    )
    plotter.plot_cluster_size_mean(summary_df)
    plotter.plot_cluster_size_max(summary_df)
    plotter.plot_noise_ratio(summary_df)

    print("Done! Check the experiments/cluster_analysis/figures folder.")
```
