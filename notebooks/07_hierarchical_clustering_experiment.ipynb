# Hierarchical Clustering Experiment

This notebook runs a hierarchical agglomerative clustering experiment on product embeddings. It builds a complete linkage tree, extracts clusters at multiple levels (k=10, 25, 50), and analyzes the semantic stability between levels using TF-IDF and Jaccard similarity.

```python
import os
import sys
import pandas as pd
import matplotlib.pyplot as plt

# Add src to path
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from src.experiments.hierarchical_clustering import HierarchicalExperiment
```

## Initialize and Run Experiment

```python
experiment = HierarchicalExperiment(
    embeddings_path="data/embeddings/product_embeddings.npy",
    dataset_path="data/raw/some_product_names.xlsx",
    text_column="Name",
    k_values=[10, 25, 50]
)

assignments_df, linkage_matrix, level_results, stability_data = experiment.run()
```

## Review Results

### Cluster Assignments Overview

```python
assignments_df[['Name', 'cluster_k10', 'cluster_k25', 'cluster_k50']].head(10)
```

### Semantic Content (Top TF-IDF Terms)

```python
for k in [10, 25, 50]:
    print(f"\nTop keywords for k={k}:")
    for cluster_id, data in level_results[k].items():
        print(f"  Cluster {cluster_id}: {', '.join(data['top_keywords'])}")
```

### Parent-Child Stability

```python
for (p, c), df_stab in stability_data.items():
    print(f"\nAverage Jaccard Similarity from k={p} to k={c}: {df_stab['jaccard_similarity'].mean():.4f}")
```
